```python
from difflib import SequenceMatcher

def find_word_replacements(words1, words2):
    # Tokenize into words
    #words1 = text1.split()
    #words2 = text2.split()
    print("words1: ", words1)
    print("words2: ", words2)
    # words1 = text1
    # words2 = text2
    # Compare sequences
    matcher = SequenceMatcher(None, words1, words2)
    opcodes = matcher.get_opcodes()
    print("opcodes: ",opcodes)
    opcodes = matcher.get_opcodes()
    # for tag, i1, i2, j1, j2 in opcodes:
    #     print('{:7}   words1[{}:{}] --> words2[{}:{}] {!r:>8} --> {!r}'.format(
    #         tag, i1, i2, j1, j2, words1[i1:i2], words2[j1:j2]))
        
    # match_opcodes = []
    # for tup in opcodes:
    #     match_opcodes.append((tup[0], words1[tup[1]:tup[2]], words2[tup[3]:tup[4]]))
    # print("opcodes words: ", match_opcodes)
    
    mapping = []
    for tag, i1, i2, j1, j2 in opcodes:
        #if tag == 'replace':
            # Extract replaced words
            replaced_words1 = words1[i1:i2]
            replaced_words2 = words2[j1:j2]
            replacements = []
            # Pair words (even if replacement lengths differ)
            max_len = max(len(replaced_words1), len(replaced_words2))
            for idx in range(max_len):
                word1 = replaced_words1[idx] if idx < len(replaced_words1) else None
                word2 = replaced_words2[idx] if idx < len(replaced_words2) else None
                replacements.append((word1, word2))
            mapping.append((tag, replacements))
    
    return mapping

# Example usage
# text1 = "Python is awesome"
#text1 = 402 - line id 1
text9 = all_spans_xcometN[1]
text10 = all_spans_xcometdefault[1]
replacements = find_word_replacements(text9, text10)
print(replacements)
print("Word replacements (text1 → text2):")
for old, new in replacements:
    print(f"'{old}' → '{new}'")
```

```python
import difflib

def token_sequence_similarity(list1, list2):
    return difflib.SequenceMatcher(None, list1, list2).ratio()

# Example
list1 = ["playing", "soccer"]
list2 = ["play", "ing", "soccer"]
similarity = token_sequence_similarity(text9, text10)
print(f"Token-level similarity: {similarity:.2f}")
```

```python
import difflib

def string_based_similarity(list1, list2):
    # Join tokens into strings (assumes subwords form original words when concatenated)
    str1 = "".join(list1)
    str2 = "".join(list2)
    return difflib.SequenceMatcher(None, str1, str2).ratio()

# Example
list1 = ["playing", "soccer"]
list2 = ["play", "ing", "soccer"]
similarity = string_based_similarity(text9, text10)
print(f"String-based similarity: {similarity:.2f}")
```
